#### (a) Types of data:

1. Numerical Data :

Numerical data consists of values that are expressed in numbers and on which arithmetic operations can be performed.

Examples:

- Age of Household Head (EX, 26, 45, 60 years)
- Household Income (EX, ₹42,000; ₹76,000; ₹91,000)
- Family Size (EX, 2, 4, 6 members)

These variables can be used to calculate mean, median, standard deviation, etc.

2. Categorical Data :

Categorical data represents characteristics or labels and cannot be measured numerically.

Examples:

- Education Level (Primary, Secondary, Graduate, Post-Graduate)
- Owns House (Yes / No)
- Urban–Rural Location (Urban, Rural)

This type of data is usually summarized using counts, percentages, and bar charts.

----------------------

#### (b) Types of Statistics

1. Descriptive Statistics :

Descriptive statistics are used to summarize, organize, and describe the main features of a dataset.

Examples:

- Average household income
- Maximum and minimum age of household heads
- Percentage of urban households

Descriptive statistics describe only the given data and do not make predictions.

2. Inferential Statistics :

Inferential statistics are used to make conclusions or predictions about a population based on sample data.

Example:

- Based on the dataset, it can be inferred that households with higher education levels tend to have higher incomes.

Inferential statistics involve estimation, hypothesis testing, and prediction.

#### (c) Descriptive Statistics
--------

Descriptive statistics refer to methods used to summarize and present data in a meaningful way.

It includes:

- Measures of central tendency (mean, median, mode)
- Measures of dispersion (range, variance, standard deviation)
- Tables, charts, and graphs

Example:
The mean household income gives the average income of all households in the dataset.

##### (a) Gaussian Distribution (Normal Distribution) :

A Gaussian distribution is a symmetrical, bell-shaped distribution where most values cluster around the mean

Explanation

- Bell-shaped curve
- Mean = Median = Mode
- Symmetrical distribution

Formula:

$$
f(x) = \frac{1}{\sigma\sqrt{2\pi}}
e^{-\frac{(x-\mu)^2}{2\sigma^2}}
$$


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

x = np.linspace(-4, 4, 100)
y = norm.pdf(x, 0, 1)

plt.plot(x, y)
plt.title("Gaussian Distribution")
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.show()
```

##### (b) Log-Normal Distribution :

A Log-Normal Distribution is a distribution in which the logarithm of the variable is normally distributed.
It is positively (right) skewed.

Explanation

- Not symmetric
- Used for income, stock prices
- Long right tail

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

x = np.linspace(0.01, 5, 100)
y = norm.pdf(np.log(x), 0, 1) / x

plt.plot(x, y)
plt.title("Log Normal Distribution")
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.show()
```

##### (c) 3-Sigma Rule (Empirical Rule) :

The 3-Sigma Rule states that in a normal distribution:

- 68% data lies within ±1σ
- 95% data lies within ±2σ
- 99.7% data lies within ±3σ

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

x = np.linspace(-3, 3, 300) 
y = norm.pdf(x)

plt.plot(x, y)
plt.axvline(-1, linestyle='--')
plt.axvline(1, linestyle='--')
plt.axvline(-2, linestyle='--')
plt.axvline(2, linestyle='--')
plt.axvline(-3, linestyle='--')
plt.axvline(3, linestyle='--')

plt.title("3-Sigma Rule")
plt.show()
```

##### Gaussian Distribution Curve

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
df = pd.read_csv("dataset.csv")


mean = df['Household_Income'].mean()
std = df['Household_Income'].std()

income_min = df['Household_Income'].min()
income_max = df['Household_Income'].max()

plt.figure(figsize=(6,4))
plt.hist(df['Household_Income'], bins=15, density=True, color='lightgreen', edgecolor='black')

x = np.linspace(income_min, income_max, 100)
y = norm.pdf(x, mean, std)
plt.plot(x, y, color='blue')

plt.title("Household Income with Gaussian Curve")
plt.xlabel("Household Income ")
plt.ylabel("Density")
plt.show()
```

#### Skewness and Kurtosis

```python
from scipy.stats import skew, kurtosis

df = pd.read_csv("dataset.csv")

skewness = skew(df['Household_Income'])
kurtosis = kurtosis(df['Household_Income'])

print(f"Skewness of Household Income: {skewness:.2f}")
print(f"Kurtosis of Household Income: {kurtosis:.2f}")
```

| Measure  | Value | Interpretation                                            |
| -------- | ----- | --------------------------------------------------------- |
| Skewness | 0.10  | Approximately symmetric, very slight right skew           |
| Kurtosis | -1.12 | Platykurtic, flatter than normal, spread out distribution |
